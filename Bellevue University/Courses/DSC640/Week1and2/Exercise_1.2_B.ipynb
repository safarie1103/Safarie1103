```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import xlrd

```

```python
plt.style.use('ggplot')
```

Load the data

```python
obama_approval_ratings = pd.read_excel('Data\obama-approval-ratings.xls')
obama_approval_ratings.head()
```

Python Bar Chart

```python
obama_approval_ratings.set_index('Issue').plot(kind='bar')
plt.xticks(rotation=90)
plt.xlabel('Issue')
plt.ylabel('Approve/Disapprove')
plt.title('Obama approval')
plt.legend(loc='center right', bbox_to_anchor=(1.3, 0.5))
plt.savefig('./SavedFigures/Python_BarChart.jpeg', dpi=200)
plt.show()
```

Python Stacked Bar Chart

```python
obama_approval_ratings.set_index('Issue').plot(kind='bar', stacked=True)
plt.xticks(rotation=90)
plt.xlabel('Issue')
plt.ylabel('Approve/disapprove')
plt.title('Obama approval')
plt.legend(loc='center right', bbox_to_anchor=(1.3, 0.5))
plt.show()
```

Python Pie Chart

```python
issues = obama_approval_ratings["Issue"]
approve = obama_approval_ratings["Approve"]


fig, ax = plt.subplots()
ax.pie(approve,  labels=issues, autopct='%1.1f%%',
        startangle=90)
ax.axis('equal') 
plt.title('Obama approval')
plt.show()

```

Python Donut Chart

```python
issues = obama_approval_ratings["Issue"]
approve = obama_approval_ratings["Approve"]


fig, ax = plt.subplots()
ax.pie(approve, labels=issues, autopct='%1.1f%%',
        startangle=90)
ax.axis('equal') 

# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
plt.title('Obama approval')
plt.show()
```

```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

df = pd.read_csv("data/costcos-geocoded.csv", delimiter=',', skiprows=0, low_memory=False)
#print(df)
#fig, ax = plt.subplots(figsize=(10,7))
#ax.set_title("Store Locations in US")
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
#print(geometry)
gdf = GeoDataFrame(df, geometry=geometry)   
#print(gdf)
#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
#print(world)
usa = world[world.name == "United States of America"]
tracePlot = gdf.plot(ax=usa.plot(figsize=(20, 20)),marker='o', color='red', markersize=15);
#fig = tracePlot.get_figure()
```

```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

df = pd.read_csv("data/costcos-geocoded.csv", delimiter=',', skiprows=0, low_memory=False)

fig, ax = plt.subplots(figsize=(10,7))
ax.set_title("Store Locations in US")
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]

gdf = GeoDataFrame(df, geometry=geometry)   

print(gdf)
#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
cities =   gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
print(cities)
#usa = world[world.name == "United States of America"]
#print(usa)
#tracePlot = gdf.plot(ax=usa.plot(figsize=(20, 20)),marker='o', color='red', markersize=15);
#fig = tracePlot.get_figure()
```
